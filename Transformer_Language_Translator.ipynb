```python
from transformers import pipeline

# Specify the model to use for translation, for example 'Helsinki-NLP/opus-mt-fr-en'
translator = pipeline('translation', model='Helsinki-NLP/opus-mt-fr-en')

# Translate text from French to English
translation = translator("Je suis étudiant")
translation
print(translation[0]['translation_text'])  # Output: I am a student
```

```python
from transformers import pipeline

# Load the translation pipeline with the specific model for French to English
translator = pipeline('translation', model='Helsinki-NLP/opus-mt-en-fr')

# Get user input in French
french_text = input("Enter the text in French for translation: ")

# Translate the text from French to English
translation = translator(french_text)

# Output the translation
print("Translated text in English:", translation[0]['translation_text'])
```

```python
from transformers import MarianMTModel, MarianTokenizer

# Load the tokenizer and model from Hugging Face's Transformers library
model_name = 'Helsinki-NLP/opus-mt-en-fr'
tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)

# Function to translate text from French to English
def translate(text):
    # Tokenize the text using the __call__ method
    model_inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True)

    # Perform the translation
    gen = model.generate(**model_inputs)

    # Decode the generated tokens to string
    translation = tokenizer.batch_decode(gen, skip_special_tokens=True)
    return translation[0]

# Example user input (for the actual use case, uncomment the input line and comment out the next one)
# user_input = input("Please enter the text in French to translate: ")
user_input = input("Enter Message") # Assuming this is the input from the user

# Translate the text and print it
translated_text = translate(user_input)
print("Translated text:", translated_text)
```
